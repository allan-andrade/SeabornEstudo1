## Os dados

Nós estaremos trabalhando com um famoso conjunto de dados do titanic.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Exercícios

Recrie os plots abaixo usando o DataFrame Titanic.

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.jointplot(x='fare',y='age',data=titanic,kind='scatter')
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.distplot(titanic['fare'],kde=False,bins=30)
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.boxplot(x="class", y="age", data=titanic,palette='rainbow')
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.swarmplot(x="class", y="age", data=titanic, size = 5)
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.countplot(x='sex',data=titanic)
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
sns.heatmap(titanic.corr(),cmap='coolwarm')
```

```python
# PONHA O SEU CÓDIGO AQUI, 
# POIS SE USASSE A LINHA DEBAIXO, PODERIA ACABAR APAGANDO O PLOT QUE DESEJAMOS ENCONTRAR.
g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age")
```
